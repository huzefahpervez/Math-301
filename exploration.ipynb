```python
# Check CSV file size in-memory
import sys
import io

# CSV size
def check_file_szie(data):
  csv_buffer = io.StringIO()
  data.to_csv(csv_buffer, index=False)
  csv_size_mb = len(csv_buffer.getvalue().encode('utf-8')) / (1024**2)
  print(f"CSV size: {csv_size_mb:.2f} MB")
```
